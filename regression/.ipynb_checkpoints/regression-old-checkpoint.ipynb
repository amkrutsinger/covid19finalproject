```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.dates as mdates
```

```python
# Load COVID-19 case data
raw_confirmed = pd.read_csv('time_series_covid19_confirmed_global.csv')
raw_deaths = pd.read_csv('time_series_covid19_deaths_global.csv')
raw_recovered = pd.read_csv('time_series_covid19_recovered_global.csv')

# Cleaning the data
confirmed = raw_confirmed.drop(['Lat','Long'], axis = 1)
deaths = raw_deaths.drop(['Lat','Long'], axis = 1)
recovered = raw_recovered.drop(['Lat','Long'], axis = 1)

# Removing province information so we have consistent country-level resolution

def set_country_res(df):

    df_sans_provinces = df.drop('Province/State', axis=1)
    df_sans_provinces = df_sans_provinces.groupby('Country/Region').sum()
    
    return df_sans_provinces

confirmed = set_country_res(confirmed)
deaths = set_country_res(deaths)
recovered = set_country_res(recovered)
```

```python
# Calculate death rate for each country
def cal_death_rate(total_deaths, total_confirmed):
    death_rate = pd.Series(dtype = float)
    
    # Calculating death rate
    if (total_deaths.index == total_confirmed.index).all():
        death_rate = total_deaths/(total_confirmed + total_deaths)
    else:
        print('Whoops, looks like your countries dont match')
    
    # Drop countries that have a null death rate (don't have any cases)
    death_rate.dropna(inplace=True)
    
    death_rate = pd.DataFrame(death_rate)
    death_rate.rename(columns={death_rate.columns[0]:'Death Rate'},inplace=True)
    
    return death_rate

# Get most recent numbers for recovered and deaths (last column in the data table)
total_confirmed_norm = confirmed[confirmed.columns[-1]]
total_deaths_norm = deaths[deaths.columns[-1]]

# Calculate death rate
death_rate_norm = cal_death_rate(total_deaths_norm, total_confirmed_norm)
death_rate_norm.head()
```

```python
# Load explanatory variables data
def load_data(file, dropCols, renameCols):
    df = pd.read_csv(file)
    df.drop(dropCols, axis=1, inplace=True)
    df.rename(columns=renameCols, inplace=True)
    df.set_index('Country/Region', drop=True, inplace=True)
    
    return df

# Load median age data
median_age = load_data('median_age.csv', 
                       ['Median Male','Median Female'],
                      {'Median':'Median Age', 'Place':'Country/Region'})
median_age.head()
```

```python
# Rename country names in data
def rename_country_names(df):
    df.rename(index={'United States':'US','South Korea':'Korea, South'}, inplace=True)

# Merge two data sets
def merge_deathrate(death_rate, data):
    df = data.merge(death_rate,left_index=True, right_index=True)
    
    # Check missing countries
    print(f'We lost {death_rate.shape[0] - df.shape[0]} countries in our merge')
    
    return df
    
rename_country_names(median_age)
list(median_age.index)

deaths_and_age = merge_deathrate(death_rate_norm, 
                                 median_age)
list(deaths_and_age.columns) 
```

```python
# Plot linear regression
def lin_regress(merged_data, col_death, col_data, line_length):
    # Linear regression using scipy
    slope, intercept, r_value, p_value, std_err = stats.linregress(merged_data[col_data], merged_data[col_death])
    predictions = np.linspace(15,line_length,5) * slope + intercept
    
    # Visualize using plt
    plt.scatter(merged_data[col_data], merged_data[col_death])
    plt.ylabel(col_death)
    plt.xlabel(col_data)
    plt.title(f'{col_death} vs {col_data}')
    plt.plot(np.linspace(15,line_length,5), predictions)
    plt.show()
    
    # Print important values
    print(f'p-values: {p_value}')
    print(f'R^2: {r_value*r_value}')
    print(f'Slope: {slope}')
    

lin_regress(deaths_and_age, 
            'Death Rate', 
            'Median Age', 
            55)
```

```python
# Task 1
# Merge total confirmed and total deaths data
filtered = pd.merge(total_confirmed_norm, total_deaths_norm, on = "Country/Region")
filtered.rename(columns={filtered.columns[0]:'Total Confirmed', filtered.columns[1]:'Total Deaths'},inplace=True)

# Filter out countries with less than min_sample_size
min_sample_size = 2000
filtered = filtered[filtered['Total Confirmed'] >= min_sample_size]


# Calculate filtered death rate
death_rate_filtered = cal_death_rate(filtered['Total Deaths'], filtered['Total Confirmed'])

# Merge filtered death rate and median age
deaths_and_age_filtered = merge_deathrate(death_rate_filtered, median_age)

# # Visualize linear regression
lin_regress(deaths_and_age_filtered, 
            'Death Rate', 
            'Median Age', 
            55)
```

```python
# Task 2: Population Density
# Load data
pop_density = load_data('population_density.csv',
                       [],
                       {'Country Name':'Country/Region', 'Value':'Pop Density'})
pop_density = pop_density[pop_density['Year'] == 2016]
pop_density.drop(['Country Code','Year'], axis=1, inplace=True)

# Merge total confirmed and total deaths data
filtered = pd.merge(total_confirmed_norm, total_deaths_norm, on = "Country/Region")
filtered.rename(columns={filtered.columns[0]:'Total Confirmed', filtered.columns[1]:'Total Deaths'},inplace=True)

# Filter out countries with less than min_sample_size
min_sample_size = 2000
filtered = filtered[filtered['Total Confirmed'] >= min_sample_size]


# Calculate filtered death rate
death_rate_filtered = cal_death_rate(filtered['Total Deaths'], filtered['Total Confirmed'])

# Merge filtered death rate and pop density
deaths_and_pop_filtered = merge_deathrate(death_rate_filtered, pop_density)

# Visualize linear regression
lin_regress(deaths_and_pop_filtered, 
            'Death Rate', 
            'Pop Density', 
            500)
```

```python
# Task 2: Death Rate by Lung Diseases
# Load data
lung_diseases = load_data('death_causes_lung_utf.csv',
                       ['Respiratory diseases', 'Trachea, bronchus, lung cancers'],
                       {'Total':'Deaths by Lung Diseases'})

pop = load_data('population.csv',
                ['LocID', 'VarID', 'Variant', 'MidPeriod', 'PopMale', 'PopFemale', 'PopDensity'],
                {'Location': 'Country/Region', 'PopTotal': 'Population'})
pop = pop[pop['Time'] == 2019]
pop.drop(['Time'], axis=1, inplace=True)

lung_diseases_pop = pd.merge(lung_diseases, pop, on = "Country/Region")
lung_diseases_pop["Death Rate by Lung-Related Diseases"] = lung_diseases_pop["Deaths by Lung Diseases"] / lung_diseases_pop["Population"]
lung_diseases_pop.drop(['Deaths by Lung Diseases', 'Population'], axis=1, inplace=True)

# Merge total confirmed and total deaths data
filtered = pd.merge(total_confirmed_norm, total_deaths_norm, on = "Country/Region")
filtered.rename(columns={filtered.columns[0]:'Total Confirmed', filtered.columns[1]:'Total Deaths'},inplace=True)

# Filter out countries with less than min_sample_size
min_sample_size = 2000
filtered = filtered[filtered['Total Confirmed'] >= min_sample_size]


# Calculate filtered death rate
death_rate_filtered = cal_death_rate(filtered['Total Deaths'], filtered['Total Confirmed'])

# Merge filtered death rate and pop density
deaths_and_lungdis_filtered = merge_deathrate(death_rate_filtered, lung_diseases_pop)
deaths_and_lungdis_filtered.describe()

def lin_regress_1(merged_data, col_death, col_data, line_length):
    # Linear regression using scipy
    slope, intercept, r_value, p_value, std_err = stats.linregress(merged_data[col_data], merged_data[col_death])
    predictions = np.linspace(2,0.1,5) * slope + intercept
    
    # Visualize using plt
    plt.scatter(merged_data[col_data], merged_data[col_death])
    plt.ylabel(col_death)
    plt.xlabel(col_data)
    plt.title(f'{col_death} vs {col_data}')
    plt.plot(np.linspace(2,0.1,5), predictions)
    plt.show()
    
    # Print important values
    print(f'p-values: {p_value}')
    print(f'R^2: {r_value*r_value}')
    print(f'Slope: {slope}')

# Visualize linear regression
lin_regress_1(deaths_and_lungdis_filtered, 
            'Death Rate', 
            'Death Rate by Lung-Related Diseases', 
            0)
```
